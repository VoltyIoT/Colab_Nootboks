#After training, optimizing model:-

```python
#  Reinstalls a specific version of paddle2onnx (v1.2.11) to ensure compatibility with your model export and optimization steps.
# - First, forcibly uninstalls any existing version of paddle2onnx to avoid version conflicts.
# - Then installs version 1.2.11, which is known to work reliably with PaddleOCR's export and ONNX conversion tools.

!python3 -m pip uninstall -y paddle2onnx
!python3 -m pip install paddle2onnx==1.2.11
```

```python
# Converts the exported PaddleOCR detection model into ONNX format using paddle2onnx.
# - model_dir: Directory containing the inference-ready Paddle model.
# - model_filename: The structure of the model (inference.pdmodel).
# - params_filename: The trained weights (inference.pdiparams).
# - save_file: Path to save the resulting ONNX model (det_db.onnx) in Google Drive.
# - opset_version: Specifies ONNX operator set version (v14 ensures compatibility with most runtimes).
# - enable_onnx_checker: Validates the exported ONNX model to ensure structural correctness.


import paddle
!paddle2onnx --model_dir /content/drive/MyDrive/PaddleOCR/inference/det_db \
--model_filename /content/drive/MyDrive/PaddleOCR/inference/det_db/inference.pdmodel \
--params_filename /content/drive/MyDrive/PaddleOCR/inference/det_db/inference.pdiparams \
--save_file /content/drive/MyDrive/PaddleOCR/inference/det_db/det_db.onnx \
--opset_version 14 \
--enable_onnx_checker True
```

```python
# Optimizes the ONNX model (det_db.onnx) for faster and more efficient inference.
# - input_model: Path to the original ONNX model exported from PaddleOCR.
# - output_model: Saves the optimized version as det_db_fixed.onnx.
# - input_shape_dict: Specifies a fixed input shape (batch size 1, 3-channel RGB, 480x480 resolution),
#   which enables graph-level optimizations and memory preallocation during runtime.

!python3 -m paddle2onnx.optimize \
  --input_model /content/drive/MyDrive/PaddleOCR/inference/det_db/det_db.onnx \
  --output_model /content/det_db_fixed.onnx \
  --input_shape_dict "{'x': [1, 3, 480, 480]}"
```
